```python
#viewing the data types
fdny = pd.read_csv("E:/jagruthi_DA/TestCases/FDNY.csv")
three_attributes = ['FacilityName','FacilityAddress', 'Borough']
data_types = fdny[three_attributes].dtypes
print(data_types)
```

```python
#select fdny information Boroughwise
grouped_data = fdny.groupby('Borough')
for Borough, Borough_data in grouped_data:
    print(f"Information of Borough:{Borough}")
    print(Borough_data)
    print("\n")
```

```python
#view fdny information for each Borough
grouped_data = fdny.groupby('Borough')
for Borough, Borough_data in grouped_data:
    print(f"Information of Borough:{Borough}")
    print(Borough_data[['FacilityName', 'FacilityAddress']])
    print(Borough_data)

    print("\n")
```
